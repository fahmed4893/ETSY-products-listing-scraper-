```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import csv
```

```python


HEADERS= {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/601.3.9 (KHTML, like Gecko) Version/9.0.2 Safari/601.3.9'}
```

```python

base_url="https://www.etsy.com/c/home-and-living/bathroom?ref=pagination&page=1"
```

```python

l=[]
page=1
for page in range (1,101):
    url=f"https://www.etsy.com/c/home-and-living/bathroom?ref=pagination&page={page}"
    response=requests.get(url, headers=HEADERS)
    soup=BeautifulSoup(response.text, "html.parser")
    #listing=soup.find_all('div',{'class':'js-merch-stash-check-listing'})
    listing=soup.find_all('li',{'class':'wt-grid__item-md-4'})
    
    #print(listing)
    
    
    for list in listing:
        
    
        information={}
    
        #title=list.find('h2',{'class':'v2-listing-card__title'}).get_text().strip()
        #information['Title']=title

        product_title=list.select_one("li.wt-grid__item-md-4 div.js-merch-stash-check-listing a.listing-link")['title'] 
        information['Title']=product_title

        #image=list.select_one("li.wt-grid__item-md-4 div.js-merch-stash-check-listing a.listing-link div.v2-listing-card__img div.height-placeholder img.sc_gallery-1-1")['src'] 
        #information['image']=image

        image=list.select_one("div.v2-listing-card__img div.height-placeholder img.wt-width-full")['src'] 
        information['image']=image

        
    
    
        saleprice=list.find('span',{'class':'currency-value'}).get_text().strip()
        information['SalePrice']=saleprice
    


        #originalprice=list.select_one('div.n-listing-card__price').text.strip()
        #information['originalprice']=originalprice


    

        product_url=list.select_one("li.wt-grid__item-md-4 div.js-merch-stash-check-listing a.listing-link")['href'] 
        information['URL']=product_url

        

        shop_name=list.select_one("div.v2-listing-card__info p.wt-mb-xs-1").text.replace("\n"," ").strip()
        information['shop_name']=shop_name

        shop_reviews=list.select_one("div.v2-listing-card__info").text.replace("\n","").strip()
        information['shopreviews']=shop_reviews

       # shoprating=list.find('div',{'class':'black-stars'}).attr['aria-label']
        #information['shoprating']=shoprating

        rating=list.find(attrs={"name":"rating"})
        information['shoprating']=rating

        #shopname=list.find('p',{'class':'streamline-seller-shop-name__line-height'}).get_text().strip()
        #information['ShopName']=shopname

        #shoprating=list.find("div",{"class":"shop-name-with-rating"}).get_text().strip()
        #information['shoprating']=shoprating

    
        l.append(information)

    print(l) 






#listing=soup.find_all('div',{'class':'js-merch-stash-check-listing'})
#listing=soup.find_all('div',{'class':'v2-listing-card__info'})
#print(listing)


```

```python
df=pd.DataFrame(l,columns=['Title','SalePrice','shop_name','shoprating','shopreviews','image','URL'])
df.head(100)                
```

```python
csv=df.to_csv('etsy_bathroom_listing_with_image.csv', index=False)
```
